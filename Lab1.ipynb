```python
class Node:
    def __init__(self, data):
        self.Data = data
        self.Prev = None
        self.Next = None

class LinkedList:
    def __init__(self):
        self.Head = None
        self.Tail = None

    # Add node at the end of the linked list
    def Add(self, data):
        nd = Node(data)
        if self.Head is None:  # Empty linked list
            self.Head = nd
            self.Tail = nd
        else:
            nd.Prev = self.Tail
            self.Tail.Next = nd
            self.Tail = nd

    # Insert a node at a specific location (0-based index)
    def Insert(self, data, loc):
        nd = Node(data)
        if self.Head is None:  # Empty linked list
            self.Head = nd
            self.Tail = nd
            return

        if loc == 0:  # Insert at head
            nd.Next = self.Head
            self.Head.Prev = nd
            self.Head = nd
            return

        cur = self.Head
        i = 0
        while i < loc - 1 and cur.Next is not None:
            cur = cur.Next
            i += 1

        if cur == self.Tail:  # Insert at tail
            nd.Prev = self.Tail
            self.Tail.Next = nd
            self.Tail = nd
        else:  # Insert in middle
            nd.Next = cur.Next
            nd.Prev = cur
            cur.Next.Prev = nd
            cur.Next = nd

    # Search for a node by data
    def Search(self, data):
        nd = self.Head
        while nd is not None:
            if nd.Data == data:
                return nd
            nd = nd.Next
        return None



    # Display all elements in the linked list
    def DisplayAll(self):
        cur = self.Head
        while cur is not None:
            print(cur.Data)
            cur = cur.Next
        print("all data")

    # Delete all nodes in the linked list
    def DeleteAll(self):
        while self.Head is not None:
            temp = self.Head
            self.Head = self.Head.Next
            del temp
        self.Tail = None  # tail not refernce to anything
  
```

```python
dll = LinkedList()
dll.Add(10)
dll.Add(20)
dll.Add(30)
dll.Insert(15, 1)
print("Linked List:")
dll.DisplayAll()  
print("\nAfter Deleting 20:")
dll.DisplayAll()  
dll.DeleteAll()
print("\nAfter Deleting All:")
dll.DisplayAll()
```
